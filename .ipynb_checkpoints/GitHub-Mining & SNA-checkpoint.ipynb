```python
def createContrib(s,u):
    contributions = [ s for s in repo.get_commits(since=s,until=u)]
    contributors=[s.author for s in contributions]
    contributors=list(dict.fromkeys(contributors))
    #Some commits belong to users that don't have an account anymore, or changed it so when inquire the user would be NonType we have to eliminate those.
    contributors=[s for s in contributors if s!=None ]
    return(contributors)
```

```python
def createFollwDict(contributors):
    from_initial_followship_dict={}
    to_initial_followship_dict={}
    for user in contributors:
        print('Contributor : ',user.login)
        from_initial_followship_dict[user.login]=[follower for follower in user.get_followers()]
    for user in contributors:
        print('Contributor : ',user.login)
        to_initial_followship_dict[user.login]=[followed for followed in user.get_following()]
    return(from_initial_followship_dict,to_initial_followship_dict)
```

```python
def createFiArray(fromFollDict,toFollDict,date):
    l=np.ndarray(shape=(0,4),dtype=str)
    i=0
    j=0
    #Followers of contributors:
    for contributor in fromFollDict.keys():
        for follower in fromFollDict[contributor]:
            l=np.append(l,[[date,contributor,follower.login,"contributer"]],axis=0)
        i+=1
    print(i)
    #The followed by contributors:
    for contributor in toFollDict.keys():
        for followed in toFollDict[contributor]:
            if followed.login in toFollDict.keys():
                l=np.append(l,[[date,followed.login,contributor,"contributer"]],axis=0)
            else:
                l=np.append(l,[[date,followed.login,contributor,"non-contributer"]],axis=0)
        j+=1
    print(j)
    print('One month is done!')
    return(l)
```

### Very important note:

You must note that this dataframe as designed and as I choosed to create the Graph later on reflects my sole interest in the contributors of the 'tensorflow' repository, so you might find nodes in the graph representing non-contributing individuals but you'll never see them receiving an edge, in other words, in-degree for these users is 0. This conception came purely from the fact that I'm intrested in the social status of these contributors within the software developpement community, as 'prestigious' members of it, and also I wnat to uncover their interest in each other (Other contributors). Studying the interests of these contributors is another subject for future projects. 

As you can guess by now that we saved two dataframes, the first (followship_data_frame) contains for each month the contributors and their followers. We must underline the fact that if a contributor has no contriibutors then you will not find him in this data frame, this might sound as a loss of information (since he/she is a contibutor and maybe a huge one) but in fact, an isolated node with no connections doesn't offer any information in an interest graph. The second dataframe (followship_data_frame1) has another phenomenon going on, since we drop duplicate relations between users, the size of relationships would monotonically decrease each month (Unless a large group of new contributors appear out of nowhere to contribute each month which is unrealistic), since mostly the same group of users contribute each month, if we happen to select the number of contributors that contributed in the Feb month for example from the second dataframe, it would be much less than the number of contributors we extracted initially. To give it an exact definition I would formulated as follows: 

**Number of contributors extracted for month x = X + Y + Z - T**

X = Number of contributors extracted from 2nd df.

Y = Number of contributors from month x with no followers.

Z = Number of contributors who contributed in previous months.

T = Nmber of users that has no followers and contributed on previous months.

```python
#Let's create this graph:
Gpool=[]
for i in range (0,len(followship_data_frame1)):
    if (followship_data_frame1.loc[followship_data_frame1.index==i,'To is contributor?'].get_values()[0]=='contributer'):
        nodeF=followship_data_frame1.loc[followship_data_frame1.index==i,'From'].get_values()[0]
        nodeT=followship_data_frame1.loc[followship_data_frame1.index==i,'To'].get_values()[0]
        Gpool.append(nodeF)
        Gpool.append(nodeT)
        edge=(nodeF,nodeT)
        G.add_edge(*edge)
```

```python
#Verification we didn't use same node multiple times.
print(len(G.node))
print(len(list(dict.fromkeys(Gpool))))
```

```python
nx.write_gml(G, 'followshipnetwork.gml') #saving the graph as a GML file (Graphe markup language)
```

The visualization is a good thing, but to make this study more intresting and mathematical, we should calculate some metrics that give us insights on the network we are studying. This field of study is called **SNA (Social network analysis)**, we will start by describing the topology of each network via a number of structural metrics that will eventually help us understand the relationships between the actors of each network.   

The topology analysis can be viewed from 3 major points of view: 

**1- Macro analysis: focuses on the general pattern of the network.**

**2- Relationship analysis: focuses on the connectivity of the network.**

**3- Micro analysis: focuses on the entities of the network (nodes,edges)**

We will start by the calculation of a basic graph metric which is the: IN/OUT Degree since our graphs are directed.

These metrics show how a particular individual is connected to other actors of the network, this connectivity can be seen as the intrest of other actors in a given user (Indegree), or how much a user is interested in other actors in the network (Outrdegree).

```python
def getDegrees(Gmonth,MCont):
    Degrees=pd.DataFrame(columns=['login','In','Out'])
    index=0
    for cont in MCont:
        Degrees.loc[index]=[cont,Gmonth.in_degree(cont),Gmonth.out_degree(cont)]
        index+=1
    return(Degrees)
```
